```python
deep_red = '#f7050e'
moderate_red="#f27e82"
red='#f5585e'
light_red='#f7b7b9'
deep_blue = '#0282fa'
blue='#2894fa'
light_blue='#48bbfa'
moderate_blue='#05a2f7'
```

```python
pat=[]
contact=[]
color=[]

for i in range (count):
  source = df.iloc[i][2]
  gender=df.iloc[i][0]
  p=df.iloc[i][1] 
  temp=[]
  for t in PatId:
    if(t in source and t!=p and t!="P"):
      temp.append(t)
  if(gender=='M'):
    color.append(red)
  else:
    color.append(blue)
  for j in range(len(temp)):
    pat.append(p)
    contact.append(temp[j])
    
```

```python
pat100=[]
contact100=[]
for i in range(len(pat_age_pair)):
    temp = pat_age_pair[i]
    temp = temp.split(',')
    suff = temp[0]
    suff = suff[3:]
    if(int(suff)<1000):
        pat100.append(pat_age_pair[i])
        contact100.append(contact_age_pair[i])
```

```python
data100 = pd.DataFrame(list(zip(pat100, contact100)),
               columns =['PatId', 'Contacts'])
```

```python
data100.head(5)
```

```python
from pyvis.network import Network 
```

```python

people = pat100+contact100
people = set(people)
people = list(people)
```

```python
df.head(5)
```

```python
colors=[]
for pp in people:
    temp = pp[1:]
    p=""
    for char in temp:
        if(char==','):
            break
        else:
            p+=char
    for i in range(count):
        if(df.iloc[i][1]==p):
            age = df.iloc[i][3]
            if(df.iloc[i][0]=='M'):
                if(age<18):
                    colors.append(light_red)
                elif (age>=18 and age<=30):
                    colors.append(moderate_blue)
                elif(age>30 and age<=60):
                    colors.append(red)
                else:
                    colors.append(deep_red)
            else:
                
                if(age<18):
                    colors.append(light_blue)
                elif (age>=18 and age<=30):
                    colors.append(moderate_blue)
                elif(age>30 and age<=60):
                    colors.append(blue)
                else:
                    colors.append(deep_blue)
                
    
```

```python
print(len(colors))
```

```python
graph = Network('1200px','1500px')
```

```python
graph.add_nodes(people,color=colors)
```

```python
src = data100.values.tolist()
for i in src:
  for j in i:
    if(j[0]!='P'):
      src.remove(i)
      break
```

```python
graph.add_edges(src)
```
